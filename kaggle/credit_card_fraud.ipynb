```python
import pandas as pd
import numpy as np

from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_validate, KFold
from sklearn.metrics import classification_report

import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.ensemble import GradientBoostingClassifier

import seaborn as sns

sns.set_style("whitegrid")
sns.set_context("poster")
data_input = 'data/creditcard.csv'
```

```python
#normalize amount
scaler = StandardScaler()
scaler.fit(df_data['Amount'].values.reshape(-1,1))
df_data['amt_normalized'] = df_data['Amount'].apply(scaler.transform)
df_data['amt_normalized'] = df_data['amt_normalized'].apply(lambda x: x[0][0])
df_data.head(2)
```

```python
def get_top_features(percent):
    '''
    uses PCA to calculate total variance of each feature. Uses that variance to return 
    the top features representing x percent of total variance in this dataset
    percent must be in range [0,1]
    '''
    if not 0 <= percent <= 1:
        print 'bad arg'
        return list()
    
    #calculate features using pca, and total variance of this dataset
    top_features = []
    df_features = df_data.drop(['Time','Amount','Class'], axis=1)
    pca = PCA(n_components = df_features.shape[1] )
    pca.fit_transform(df_features)
    variance_to_features = sorted(zip(pca.explained_variance_, df_features.columns), reverse=True)
    total_variance = sum([i[0] for i in variance_to_features])

    #get top features while the total variance accounted for is less than x percent of total
    variance_accounted = 0
    for var, feature in variance_to_features:
        variance_accounted += var
        top_features.append(feature)
        if variance_accounted > total_variance * percent:
            break
    return top_features

top_features = get_top_features(0.9)   
top_features
    
    
```
